# Word Clouds

A word cloud is an easy and convenient way to summarize and visualize written data.  

One of the reasons this is so easy and convenient is that someone wrote a module to do most of the hard work.  This is a good opportunity for me to show you this `try-except` trick.  If the `wordcloud` module is not installed, when you try to import it, you'll get an error.  However, since it's put inside a `try`, when it fails then the module is installed using `!pip install`.  

The first time you run this next cell, it will probably take a long time and you might see some messages about the module being downloaded and installed.  The second time you run it, it will run faster and no messages will follow it.  


```python
try:
    from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
except:
    ! pip install wordcloud
    from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
    
import numpy as np
import pandas as pd
#from os import path
#from PIL import Image

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

import warnings
warnings.filterwarnings("ignore")

```

```python
bf = pd.read_csv("bigfoot_descriptions.csv", index_col = 0)

bf
```

```python
# Start with one review:
text = bf.observed[0]

# Create and generate a word cloud image:
wordcloud = WordCloud(background_color="white").generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from urllib.request import urlopen 
import re
def read_url(url): 
    return re.sub('\\s+', ' ', urlopen(url).read().decode())

huck_finn_url = 'https://www.inferentialthinking.com/data/huck_finn.txt'
huck_finn_text = read_url(huck_finn_url)

little_women_url = 'https://www.inferentialthinking.com/data/little_women.txt'
little_women_text = read_url(little_women_url)
```

```python
wordcloud = WordCloud(background_color="white", colormap="rainbow").generate(little_women_text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
DeclarationofIndependence = read_url('https://cdn.serc.carleton.edu/files/teaching_computation/workshop_2018/activities/plain_text_version_declaration_inde.txt')

wordcloud = WordCloud(background_color="white", colormap ="flag").generate(DeclarationofIndependence)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(background_color="white", colormap ="flag_r").generate(DeclarationofIndependence)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
